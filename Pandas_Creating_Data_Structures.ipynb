# Pandas Tutorial: Creating Pandas Data Structures
In this tutorial, we will learn about how to create these data structures.
<br> Pandas has three main data structures:
- **Series**: A one-dimensional labeled array capable of holding any data type. A series is similar with one-dimensional NumPy array, but indexed. 
- **DataFrame**: A two-dimensional labeled array with columns of potentially different types. A dataframe can be seen as an indexed two-dimensional NumPy array. It is similar to a spreadsheet or SQL table. 
- **Panel**: A three-dimensional data container.

All Pandas data structures are value mutable. 
<br>In this tutorial, we focus on series and dataframe, which are the most commonly used data structures.

## 1.0. Creating a series <a id="1.0."></a>
The official documentation of Pandas series can be found __[here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.Series.html)__ (version 0.23.4). 
<br>A series can be created in three ways:
- an array/list
- a Python dict
- a scalar value

```python
#Importing NumPy module and aliasing as np.
import numpy as np

#Importing Pandas module and aliasing as pd.
import pandas as pd
```

### 1.1. Creating a series from an array/list <a id="1.1."></a>

#### Example 1 <a id="1.1.1. Example 1"></a>
Suppose we want to create a series that contains a set of integers [1, 2, 3, 4, 5]. We can do this by calling:

```python
# Creating variable 's' that contains a set of integers [1,2,3,4,5]. 
s = pd.Series([1,2,3,4,5])

# Showing the 's' series.
s
```

A series is indexed, so Pandas automatically assigns index to the data if we do not specify index in the first place. In this case, the index will be automatically set as [0,..., (array length) -1]. We can call the index parameter if we want to assign specific index to the values as shown as follows: 

```python
# Creating variable 't' that contains a set of integers [1,2,3,4,5], indexed with ['A','B','C','D','E'].
t = pd.Series([1,2,3,4,5], index=['A','B','C','D','E'])

# Showing the output of 't'.
t
```

#### Example 2 <a id="1.1.1.Example2"></a>
Suppose we want to create a shopping list, with the index be the name of the items and the values be the number of items that need to be bought, as shown below. 

```python
from IPython.display import Image
Image(filename='Table_Example3.png', width=150)
```

```python
# Creating 'shoppingList' series
shoppingList = pd.Series(data=[5, 10, 2, 1, 7], index=['Banana', 'Apple', 'Onion', 'Potato', 'Carrot'])

# Showing the output of 'shoppingList'.
shoppingList
```

The example above can also be written as follow:

```python
# Creating 'shoppingList' series.
shoppingList = pd.Series(data=np.array([5, 10, 2, 1, 7]), index=np.array(['Banana', 'Apple', 'Onion', 'Potato', 'Carrot']))

# Showing the output of 'shoppingList'.
shoppingList
```

Python list can also be used as the input for creating a Pandas series

```python
# Assigning 'index' variable the name of the shopping items.
index = ['Banana', 'Apple', 'Onion', 'Potato', 'Carrot']

# Assigning 'number' variable the number of items.
number = [5, 10, 2, 1, 7]

# Creating 'shoppingList' series based on the 'index' and 'number' variable. 
shoppingList = pd.Series(data=number, index=index)

# Showing the output of 'shoppingList'.
shoppingList
```

### 1.2. Creating a series from a Python dictionary <a id="1.2."></a>

#### Example 1
A Python dictionary can be used to initiate a Pandas series. If no index specified, the dictionary keys will become the series index. The dictionary keys are taken in a sorted order to construct index. To illustrate this, the shopping list [example](#1.1.1.Example2) is taken as follows: 

```python
# Creating 'shoppingList_dict' dictionary.
shoppingList_dict = {'Banana':5, 'Apple':10, 'Onion':2, 'Potato':1, 'Carrot':7}

# Creating 'shoppingList_series' based on 'shoppingList_dict' dictionary.
shoppingList_series = pd.Series(shoppingList_dict)

# Showing the output of 'shoppingList_series'.
shoppingList_series
```

If an index is passed, the values are ordered based on the specified index. 

```python
# Creating 'shoppingList_dict' dictionary.
shoppingList_dict = {'Banana':5, 'Apple':10, 'Onion':2, 'Potato':1, 'Carrot':7}

# Creating 'index' variable to be the index of 'shoppingList_series'.
index = np.array(['Apple', 'Potato', 'Onion', 'Carrot', 'Banana'])

# Creating 'shoppingList_series' based on 'shoppingList_dict' dictionary.
shoppingList_series = pd.Series(shoppingList_dict, index=index)

# Showing the output of 'shoppingList_series'
shoppingList_series
```

If an index is passed, the values in data corresponding to the labels in the index will be pulled out. Therefore, if the index contains any label that the dictionary keys do not have, it will return missing value (NaN). NaN (not a number) is the standard missing data marker used in pandas.

```python
# Creating 'shoppingList_dict' dictionary.
shoppingList_dict = {'Banana':5, 'Apple':10, 'Onion':2, 'Potato':1, 'Carrot':7}

# Creating 
index = np.array(['Orange', 'Potato', 'Onion', 'Carrot', 'Banana'])

# Creating 'shoppingList_series' based on 'shoppingList_dict' dictionary.
shoppingList_series = pd.Series(shoppingList_dict, index=index)

# Showing the output of 'shoppingList_series'.
shoppingList_series
```

### 1.3. Creating a series from a scalar value <a id="1.3."></a>

#### Example 1
If data is a scalar value, an index must be provided. The value will be repeated to match the length of index.

```python
# Creating variable 'v' to initiate a series with scalar value of 8. 
v = pd.Series(8, index=[0,1,2,3,4])

# Showing the output of 'v'.
v
```

#### Example 2
A string can also be used as a scalar value. 

```python
# Creating variable 'v' to initiate a series with scalar value of 'Apple. 
v = pd.Series('Apple', index=[0,1,2,3,4])

# Showing the output of 'v'.
v
```

## 2.0. Creating a dataframe <a id="2.0."></a>
The official documentation of Pandas series can be found __[here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.DataFrame.html)__ (version 0.23.4). 
<br> A dataframe can be created by using various input, such as:
- a nested list/array
- a dictionary of lists/arrays
- a dictionary of series
- a list of dictionaries

### 2.1.  Creating a dataframe from a nested list/array <a id="2.1."></a>

#### Example 1 <a id="1.2.1.Example1"></a>
A nested list/array is a list/array that appears as an element in another list/array. Suppose we want to create a dataframe containing name, age, and occupation of personnel as shown below.

```python
from IPython.display import Image
Image(filename='Table_Example1.png', width=220)
```

We can use a nested list to do this. 

```python
# Creating 'data' variable, containing personnel information, by using a nested list
data = [['Dario', 29, 'Farmer'], ['Andreas', 35, 'Doctor'], ['Tim', 19, 'Student'], ['Mike', 47, 'Engineer']]

# Creating 'identity' variable based on the 'data' variable
identity = pd.DataFrame(data, index=[1,2,3,4], columns=['Name', 'Age', 'Occupation'])

# Showing the 'identity' dataframe.
identity
```

The same dataframe can be created by using a nested array. 

```python
# Creating 'data' variable, containing personnel information, by using nested array
data = np.array([['Dario', 29, 'Farmer'], ['Andreas', 35, 'Doctor'], ['Tim', 19, 'Student'], ['Mike', 47, 'Engineer']])

# Creating 'identity' variable based on the 'data' variable
identity = pd.DataFrame(data, index=[1,2,3,4], columns=['Name', 'Age', 'Occupation'])

# Showing the 'identity' dataframe.
identity
```

As you can see above, the inner arrays/lists of the nested array/list become the rows of the dataframe. 

#### Example 2 <a id="1.2.1.Example2"></a>
Suppose we want to record profit and revenue of a company from 2012-2016 as shown below.

```python
Image(filename='Table_Example2.png', width=220)
```

We can do this by using the following.

```python
# Creating 'data' variable, containing proft and revenue of a company, by using nested array
data = np.array([[1000, 2000], [1010, 2000], [500, 1000], [200, 300], [1500, 3000]])

# Creating 'profitRevenue' variable based on the 'data' variable
profitRevenue = pd.DataFrame(data, index=[2012, 2013, 2014, 2015, 2016], columns=['Profit', 'Revenue'])

# Showing the 'identity' dataframe.
profitRevenue
```

### 2.2. Creating a dataframe from a dictionary of lists/arrays <a id="2.2."></a>
We can create a dataframe from a dictionary of lists/arrays. The dictionary keys, by default, are taken as the column names and the dictionary values are the column values. All the arrays/lists must have the same length. Otherwise, it will return error. 

#### Example 1
To demonstrate the creation of a dataframe from a dictionary of lists/arrays, we take the previous [example](#1.2.1.Example1), where we want to create a dataframe containing name, age, and occupation of personnel.

```python
# Creating 'data' variable, containing personnel information, by using a dictionary of lists.
data = {'Name':['Dario','Andreas','Tim','Mike'], 'Age':[29,35,19,47], 'Occupation':['Farmer','Doctor','Student','Engineer']}

# Creating 'identity' variable based on the 'data' variable
identity = pd.DataFrame(data, index=[1,2,3,4])

# Showing the 'identity' dataframe.
identity
```

The same dataframe can be created by using a dictionary of arrays.

```python
# Creating 'data' variable, containing personnel information, by using a dictionary of arrays.
data = {'Name':np.array(['Dario','Andreas','Tim','Mike']), 'Age':np.array([29,35,19,47]), 'Occupation':np.array(['Farmer','Doctor','Student','Engineer'])}

# Creating 'identity' variable based on the 'data' variable.
identity = pd.DataFrame(data, index=[1,2,3,4])

# Showing the 'identity' dataframe.
identity
```

#### Example 2
We take the previous [example](#1.2.1.Example2), where we want to record profit and revenue of a company from 2012-2016. We can do this by using a dictionary of lists.

```python
# Creating 'data' variable, containing proft and revenue of a company, by using a dictionary of lists
data = {'Profit':[1000,1010,500,200,1500], 'Revenue':[2000,2000,1000,300,3000]}

# Creating 'profitRevenue' variable based on the 'data' variable
profitRevenue = pd.DataFrame(data, index=[2012, 2013, 2014, 2015, 2016])

# Showing the 'identity' dataframe.
profitRevenue
```

### 2.3. Creating a dataframe from a dictionary of series <a id="2.3."></a>
Similar with the creation of a dataframe from a dictionary of lists/arrays, the dictionary keys by default are taken as the column names and the dictionary values are taken as the column values.

#### Example 1 
To demonstrate the creation of a dataframe from a dictionary of series, we take the previous [example](#1.2.1.Example1), where we want to create a dataframe containing name, age, and occupation of personnel.

```python
# Creating 'data' variable, containing personnel information, by using a dictionary of series.
data = {'Name':pd.Series(['Dario','Andreas','Tim','Mike']), 
        'Age':pd.Series([29,35,19,47]), 
        'Occupation':pd.Series(['Farmer','Doctor','Student','Engineer'])}

# Creating 'identity' variable based on the 'data' variable
identity = pd.DataFrame(data)

# Showing the 'identity' dataframe.
identity
```

Individual index can be passed on to each series. If the indexes are different for each dictionary key, the resultant index is the union of all the series indexes passed. To demonstrate this, we specify different index for the 'Name' column (index = [1,2,3,4]) and the 'Age' and 'Occupation' column (index = [0,1,2,3]). 

```python
# Creating 'data' variable, containing personnel information, by using a dictionary of series.
data = {'Name':pd.Series(['Dario','Andreas','Tim','Mike'], index=[1,2,3,4]), 
        'Age':pd.Series([29,35,19,47], index=[0,1,2,3]), 
        'Occupation':pd.Series(['Farmer','Doctor','Student','Engineer'], index=[0,1,2,3])}

# Creating 'identity' variable based on the 'data' variable
identity = pd.DataFrame(data)

# Showing the 'identity' dataframe.
identity
```

As you can see above, the 'Name' column does not have value for index '0', so the value returns NaN. Similarly, the 'Age' and 'Occupation' column do not have value for row '4', so they return NaN. 

#### Example 2
We take the previous [example](#1.2.1.Example2), where we want to record profit and revenue of a company from 2012-2016. We can do this by using a dictionary of series.

```python
# Creating 'data' variable, containing proft and revenue of a company, by using a dictionary of series
data = {'Profit':pd.Series([1000,1010,500,200,1500], index=[2012, 2013, 2014, 2015, 2016]), 
        'Revenue':pd.Series([2000,2000,1000,300,3000], index=[2012, 2013, 2014, 2015, 2016])}

# Creating 'profitRevenue' variable based on the 'data' variable
profitRevenue = pd.DataFrame(data)

# Showing the 'identity' dataframe.
profitRevenue
```

### 2.4. Creating a dataframe from a list of dictionaries <a id="2.4."></a>
When creating a dataframe for a list of dictionaries, the dictionary keys are by default taken as column names.

#### Example 1
To demonstrate the creation of a dataframe from a list of dictionaries, we take the previous [example](#1.2.1.Example1), where we want to create a dataframe containing name, age, and occupation of personnel.

```python
# Creating 'data' variable, containing personnel information, by using a list of dictionaries.
data = [{'Name':'Dario', 'Age':29, 'Occupation':'Farmer'}, 
        {'Name':'Andreas', 'Age':35, 'Occupation':'Doctor'},
        {'Name':'Tim', 'Age':19, 'Occupation':'Student'},
        {'Name':'Mike', 'Age':47, 'Occupation':'Engineer'}]

# Creating 'identity' variable based on the 'data' variable
identity = pd.DataFrame(data, index=[1,2,3,4])

# Showing the 'identity' dataframe.
identity
```

#### Example 2
We take the previous [example](#1.2.1.Example2), where we want to record profit and revenue of a company from 2012-2016. We can do this by using a list of dictionaries.

```python
# Creating 'data' variable, containing proft and revenue of a company, by using a list of dictionaries
data = [{'Profit':1000, 'Revenue':2000}, 
        {'Profit':1010, 'Revenue':2000},
        {'Profit':500, 'Revenue':1000},
        {'Profit':200, 'Revenue':300},
        {'Profit':1500, 'Revenue':3000}]

# Creating 'profitRevenue' variable based on the 'data' variable
profitRevenue = pd.DataFrame(data, index=[2012, 2013, 2014, 2015, 2016])

# Showing the 'identity' dataframe.
profitRevenue
```
